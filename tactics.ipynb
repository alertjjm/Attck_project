```python
import pandas as pd
import json
```

```python
f=open("linux_tactics.json",'r')
jsondata=json.load(f)
f.close()
```

```python
tactickeys=jsondata.keys()
tactics=[]
for i in tactickeys:
    tactics.append(i)
```

```python
maxlen=0
for i in tactics:
    if len(jsondata[i])>maxlen:
        maxlen=len(jsondata[i])
print(maxlen)
```

```python
for i in tactics:
    while len(jsondata[i])!=maxlen:
        jsondata[i].append(" ")
```

```python
df=pd.DataFrame(jsondata,columns=tactics)
```
